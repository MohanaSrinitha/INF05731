# **Question 1: Topic Modeling**

(30 points). This question is designed to help you develop a feel for the way topic modeling works, the connection to the human meanings of documents. Based on the dataset from assignment three, write a python program to **identify the top 10 topics in the dataset**. Before answering this question, please review the materials in lesson 8, especially the code for LDA, LSA, and BERTopic. The following information should be reported:

1. Features (text representation) used for topic modeling.

2. Top 10 clusters for topic modeling.

3. Summarize and describe the topic for each cluster.


```python
# Write your code here

# Installing the necessary libraries
!pip install gensim
!pip install nltk

# Downloading the NLTK resources
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

import pandas as pd
from gensim import corpora
from gensim.models import LdaModel
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

# Loading the dataset from previous assignment
data_frame = pd.read_csv('data_reviews.csv')

#1.Features (text representation) used for topic modeling
# performing the tokenization and preprocessing
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
#defining a function to preprocess the text
def preprocess(text):
    tokens = word_tokenize(text.lower())
    tokens = [lemmatizer.lemmatize(token) for token in tokens if token.isalpha() and token not in stop_words]
    return tokens

# Identifying the text column
text_column = data_frame.select_dtypes(include=['object']).columns[0]  # Assumes the text data is in the first object column
data_frame['tokens'] = data_frame[text_column].apply(preprocess)

# Creating a dictionary to representation of the documents.
dictionary = corpora.Dictionary(data_frame['tokens'])

# Creating a corpus from the dictionary
corpus = [dictionary.doc2bow(tokens) for tokens in data_frame['tokens']]

# Building the LDA model
lda_model = LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)

print("\n2.Top 10 clusters for topic modeling")
# Printing out the top 10 topics
for topic_id, words in lda_model.print_topics(num_topics=10):
    print(f"Topic {topic_id + 1}: {words}")

# Assigning the topics to each document
data_frame['topic'] = data_frame['tokens'].apply(lambda x: max(lda_model[dictionary.doc2bow(x)], key=lambda y: y[1])[0])

print("\n3.Summarize and describe the topic for each cluster")
# Finally Displaying the top 10 clusters and their descriptions
top_clusters = data_frame['topic'].value_counts().head(10)
for cluster_id, count in top_clusters.items():
    print(f"\nCluster {cluster_id + 1} - Count: {count}")
    print(data_frame[data_frame['topic'] == cluster_id][text_column].head())
```

```python

# Write your code here
#Importing the libraries
import pandas as pd
df=pd.read_csv('data_reviews.csv')
print(df.shape) #printing out the shape of dataframe
df.head() #printing out the first 5 columns from dataset
df['sentiment'].value_counts() #checking the frequency of sentiments
```

**Here from the data_reviews.csv file taken from previous assignmnet we have three columns among which I have considered the two columns clean_text and sentiment as they are mostly required here for the sentiment analysis.**

```python
#importing the necessary libraries
import nltk
#cleaning the text
df['clean_text'] = df['clean_text'].str.replace('[^\w\s]','')  #removal of punctuation
df['clean_text'] = df['clean_text'].apply(lambda x: " ".join(x.lower() for x in x.split()))#lower case
from nltk.corpus import stopwords #removal of stopwords
nltk.download('stopwords')
stop = stopwords.words('english')
df['clean_text'] = df['clean_text'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df.head() #displaying few columns of the dataset
```

```python
df.info() #displaying the dataset information
```

```python
#importing TfidfVectorizer class from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer
#initializing a TfidfVectorizer object
tf_idf= TfidfVectorizer(ngram_range=(1,2), max_features=1000)
tf_idf.fit(df['clean_text'])
x_values =  tf_idf.transform(df['clean_text'])
y_values = df['sentiment']

# Splitting the training data to training and validating data
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(x_values, y_values, test_size = 0.2)
```

```python
#importing metrics from sklearn
from sklearn import metrics
#defining a function to calculate the required metrics
def evaluation(y_pred, y_test):
    Accuracy = metrics.accuracy_score(y_pred, y_test)
    Recall = metrics.recall_score(y_pred = y_pred, y_true = y_test, pos_label='positive', average='micro') # micro calculates total true positives, false negatives and false positives
    Precision = metrics.precision_score(y_pred = y_pred, y_true = y_test, pos_label='positive', average='micro') # micro calculates total true positives, false negatives and false positives
    F1 = 2 * (Precision * Recall) / (Precision + Recall) # Formula for F1 Score
    print("Accuracy: ", Accuracy.round(4))
    print("Recall:", Recall.round(4))
    print("Precision:", Precision.round(4))
    print("F-1 score:", F1.round(4))
```

```python
#Here I am using naive bayes and svm as the supervised learning algorithms to build sentiment classifiers  and to calculate the metrics
#first I am building the classifier with Naive bayes
#importing the libraries and modules
from sklearn import naive_bayes
from sklearn.model_selection import cross_val_score, KFold

naive_bayes_implement = naive_bayes.MultinomialNB()
naive_bayes_implement.fit(x_train,y_train)

y_pred_valid = naive_bayes_implement.predict(x_valid)
evaluation(y_pred_valid, y_valid) #calling the above defined function to perform the metrics calculation for Naive bayes
print("\nNaive Bayes")
print("Cross Validation Score:", cross_val_score(naive_bayes_implement, x_valid, y_valid, cv= KFold(10, shuffle=True, random_state = 22)))
```

```python
#importing the libraries and modules
from sklearn import svm

svm_implement = svm.SVC()
svm_implement.fit(x_train,y_train)

y_pred_valid = svm_implement.predict(x_valid)
evaluation(y_pred_valid, y_valid)#calling the above defined function to perform the metrics calculation for SVM
print("Cross Validation Score:", cross_val_score(svm_implement, x_valid, y_valid, cv= KFold(10, shuffle=True, random_state = 22)).mean().round(4))

```

```python
#importing the required libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import RobertaTokenizer, RobertaForSequenceClassification, AdamW
from torch.utils.data import DataLoader, TensorDataset
import torch
from sklearn.metrics import classification_report

# Loading the dataset
df1 = pd.read_csv('data_reviews.csv')

# Splitting the data into training and validation/test sets
train_data, test_data, train_labels, test_labels = train_test_split(df1['clean_text'], df1['sentiment'], test_size=0.2, random_state=42)

# Loading the RoBERTa tokenizer and model
tokenizer_roberta = RobertaTokenizer.from_pretrained('roberta-base')
roberta_model = RobertaForSequenceClassification.from_pretrained('roberta-base', num_labels=len(df1['sentiment'].unique()))

# Tokenize and prepare input data
train_encodings_roberta = tokenizer_roberta(train_data.tolist(), truncation=True, padding=True, max_length=128, return_tensors='pt')
test_encodings_roberta = tokenizer_roberta(test_data.tolist(), truncation=True, padding=True, max_length=128, return_tensors='pt')

train_dataset_roberta = TensorDataset(train_encodings_roberta['input_ids'], train_encodings_roberta['attention_mask'], torch.tensor(pd.factorize(train_labels)[0]))
test_dataset_roberta = TensorDataset(test_encodings_roberta['input_ids'], test_encodings_roberta['attention_mask'], torch.tensor(pd.factorize(test_labels)[0]))

# Create DataLoader for efficient batch processing
train_loader_roberta = DataLoader(train_dataset_roberta, batch_size=8, shuffle=True)
test_loader_roberta = DataLoader(test_dataset_roberta, batch_size=8, shuffle=False)

```

```python
# Setting up GPU if available
device_roberta = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
roberta_model.to(device_roberta)

# Setting up training parameters
optimizer_roberta = AdamW(roberta_model.parameters(), lr=2e-5)
num_epochs_roberta = 3
```

```python
# Training loop
for epoch_roberta in range(num_epochs_roberta):
    roberta_model.train()
    for batch_roberta in train_loader_roberta:
        input_ids_roberta, attention_mask_roberta, labels_roberta = batch_roberta
        input_ids_roberta, attention_mask_roberta, labels_roberta = input_ids_roberta.to(device_roberta), attention_mask_roberta.to(device_roberta), labels_roberta.to(device_roberta)

        optimizer_roberta.zero_grad()
        outputs_roberta = roberta_model(input_ids_roberta, attention_mask=attention_mask_roberta, labels=labels_roberta)
        loss_roberta = outputs_roberta.loss
        loss_roberta.backward()
        optimizer_roberta.step()
```

```python
# Validation loop
roberta_model.eval()
predictions_roberta = []
true_labels_roberta = []

with torch.no_grad():
    for batch_roberta in test_loader_roberta:
        input_ids_roberta, attention_mask_roberta, labels_roberta = batch_roberta
        input_ids_roberta, attention_mask_roberta, labels_roberta = input_ids_roberta.to(device_roberta), attention_mask_roberta.to(device_roberta), labels_roberta.to(device_roberta)

        outputs_roberta = roberta_model(input_ids_roberta, attention_mask=attention_mask_roberta)
        logits_roberta = outputs_roberta.logits
        predictions_roberta.extend(torch.argmax(logits_roberta, dim=1).cpu().numpy())
        true_labels_roberta.extend(labels_roberta.cpu().numpy())


# Printing the lengths of true_labels_roberta and predictions_roberta
print(len(true_labels_roberta), len(predictions_roberta))

# Printing a sample of true_labels_roberta and predictions_roberta
print("True labels:", true_labels_roberta[:10])
print("Predicted labels:", predictions_roberta[:10])

# Printing classification report or any other evaluation metrics of interest
print(classification_report(true_labels_roberta, predictions_roberta))

```
